```python
# trending video with most views
Max_views=data_US.sort_values(by='view_count',ascending=False).drop_duplicates(subset='video_id')
Max_views.head(30)
```

```python
def classify_views(element):
  if element >1000000:
    return 'above one million'
  else:
    return 'below one million'
```

```python
def classify_likes(element):
  if element >20000:
    return "above 20k"
  else:
    return 'below 20k'
```

```python
def classify_dislikes(element):
  if element >1000:
    return 'above 1k'
  if element <1000:
    return 'below 1k'
```

```python
def classify_comments(element):
  if element >1000:
    return 'above 1k'
  else:
    return 'below 1k'
```

```python
view_US=data_US['view_count'].apply(classify_views)
likes_US=data_US['likes'].apply(classify_likes)
dislikes_US=data_US['dislikes'].apply(classify_dislikes)
comments_US=data_US['comment_count'].apply(classify_comments)
```

```python
# Prepare the dataset for Machine learning
df=data_US.groupby(['video_id','trending_date','publishedAt'],as_index=False).agg({'view_count':'max','likes':'max','dislikes':'max','comment_count':'max'})
```

```python
df2 = df.copy()
df2['publishedAt']=pd.to_datetime(df2['publishedAt'])
df2['trending_date']=pd.to_datetime(df2['trending_date'])
df2['trending_day_no']=df2.groupby(['video_id'])['trending_date'].rank('first',ascending=True)
```

```python
df2_count = df2.groupby('video_id',as_index=False)["trending_date"].count().rename(columns={'trending_date':'total_trending_days'})
df3=df2.merge(df2_count,left_on='video_id',right_on='video_id')
```

```python
df3['published_year'] = df3.publishedAt.dt.year
df3['published_month'] = df3.publishedAt.dt.month
df3['published_day'] = df3.publishedAt.dt.day
df3['published_hour'] = df3.publishedAt.dt.hour
```

```python
df3.drop(columns=['video_id','trending_date','publishedAt','trending_day_no'],inplace=True)
df3.head(5)
```

```python
df3=df3.iloc[:, [4,0,1,2,3,5,6,7,8]]
```

```python
df3.corr()
```

```python
# Decision Tree
import sklearn.tree
dtree=sklearn.tree.DecisionTreeRegressor(max_depth=5)
X=df3.drop(['total_trending_days'],axis=1)
Y=df3.total_trending_days
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
dtree.fit(X_train,Y_train)
```

```python
Y_pred=dtree.predict(X_test)
# MAD
mad=(Y_pred-Y_test).abs().mean()
# MSE
mse=((Y_pred-Y_test)**2).mean()
acc = dtree.score(X_test, Y_test)
print(mad)
print(mse)
print(acc)
```
