```python
# trending video with most views
Max_views=data_US.sort_values(by='view_count',ascending=False).drop_duplicates(subset='video_id')
Max_views.head(5)
```

```python
def classify_views(element):
  if element >1000000:
    return 'above one million'
  else:
    return 'below one million'
```

```python
def classify_likes(element):
  if element >20000:
    return "above 20k"
  else:
    return 'below 20k'
```

```python
def classify_dislikes(element):
  if element >1000:
    return 'above 1k'
  if element <1000:
    return 'below 1k'
```

```python
def classify_comments(element):
  if element >1000:
    return 'above 1k'
  else:
    return 'below 1k'
```

```python
view_US=data_US['view_count'].apply(classify_views)
likes_US=data_US['likes'].apply(classify_likes)
dislikes_US=data_US['dislikes'].apply(classify_dislikes)
comments_US=data_US['comment_count'].apply(classify_comments)
```

```python
classified=pd.concat([data_US['channelTitle'],data_US['categoryId'],view_US,likes_US,dislikes_US,comments_US],axis=1)
classified
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(nrows=1,ncols=4,figsize=(16,5))
classified.groupby(['categoryId','view_count']).size().unstack().plot.bar(title="Views",ax=ax[0])
classified.groupby(['categoryId','likes']).size().unstack().plot.bar(title="Likes",ax=ax[1])
classified.groupby(['categoryId','dislikes']).size().unstack().plot.bar(title="Dislikes",ax=ax[2])
classified.groupby(['categoryId','comment_count']).size().unstack().plot.bar(title="Comments",ax=ax[3])
```
